```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from skimage import exposure  
from skimage.feature import hog
```

```python
def hog_descriptor(dataroot):

    image = cv2.imread(dataroot)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # HOG descriptor
    fd, hog_image = hog(image, orientations=8, pixels_per_cell=(28, 28), cells_per_block=(4, 4), 
                        visualize=True, channel_axis=-1)

    # Rescale histogramm for better visualization
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))

    # display images: original + descriptor
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)

    ax1.axis('off')
    ax1.imshow(image, cmap=plt.cm.gray)  
    ax1.set_title('Input image')

    ax2.axis('off')
    ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)    
    ax2.set_title('Histogram of oriented gradient')

    return
```
