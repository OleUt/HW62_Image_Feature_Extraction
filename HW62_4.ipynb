```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:/Users/MyWork/PycharmProjects/HW60_Edge_Filter/dog1.jpg')
gray= cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray
```

```python
# SIFT object
sift = cv2.SIFT_create()

# key points, descriptors
kp,des = sift.detectAndCompute(gray, None)

# put keypoints on the image
img_keypoints = cv2.drawKeypoints(gray,kp,img,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

# save and show image with keypoints
cv2.imwrite('SIFT_keypoints.jpg', img_keypoints)
plt.imshow(img_keypoints)
plt.show()
```
